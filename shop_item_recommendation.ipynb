```
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```
data=pd.read_csv('/content/drive/MyDrive/shopdata.csv',encoding= 'unicode_escape')
data.head()
```

```
data.drop(columns=['InvoiceNo', 'StockCode', 'UnitPrice', 'Country','InvoiceDate'],inplace=True)
data=data.dropna(how='any')

data.head()
```

```
buys = pd.DataFrame(data.groupby('Description')['Quantity'].mean())  
  
buys['Total Quantity'] = pd.DataFrame(data.groupby('Description')['Quantity'].count()) 
  
buys.head()
```

```
salesmat = data.pivot_table(index ='CustomerID', 
              columns ='Description', values ='Quantity') 
  
salesmat.head(15) 
  
#buys.sort_values('num of buys', ascending = False).head(10)
```

```
giftset = salesmat[' SPACEBOY BABY GIFT SET']
giftset
```

```
similar_to_giftset = salesmat.corrwith(giftset)
corr_giftset = pd.DataFrame(similar_to_giftset, columns =['Correlation'])
corr_giftset = corr_giftset.join(buys['Total Quantity'])  
corr_giftset.dropna(inplace = True)
corr_giftset.head()
```

All these items are similar to SPACEBOY BABY GIFT SET!!
